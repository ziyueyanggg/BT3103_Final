```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
df=pd.read_csv("https://docs.google.com/spreadsheets/d/14dZQ5XLFG5-1NNGYTgclrTXlkEUiSWWiOCdkjNZaAfQ/gviz/tq?tqx=out:csv&sheet=personal_data")
```

```python
df.head()
```

```python
# remove na rows
df.dropna()
```

```python
# a new dataframe to store clean data
clean=pd.DataFrame()
```

```python
clean["token"]=df["token"]
```

```python
# get age
clean["age"]= 2019-df["year_of_birth"]
```

```python
cols=df.columns.tolist()
```

```python
cols
```

```python
cols=cols[2:6]+cols[7:8]
```

```python
cols
```

```python
for i in range(0,len(cols)):
    clean=pd.concat([clean,pd.get_dummies(df[cols[i]])],axis=1)
```

```python
clean.head()
```

```python
clean.columns.tolist()
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
clean_withoutToken=clean.drop(columns="token")
```

```python
clean_withoutToken.head(1)
```

```python
kmeans.fit(clean_withoutToken)  
```

```python
cluster=kmeans.labels_.tolist()
```

```python
clean["label"]=cluster
```
